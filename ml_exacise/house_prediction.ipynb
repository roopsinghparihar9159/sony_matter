```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
area = np.array(df.area)
price = np.array(df.price)
price
```

```python
%matplotlib inline
plt.xlabel('area in sq feet')
plt.ylabel('Price in US $')
plt.scatter(df.area,df.price,color="red",marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
# lr.fit(x.reshape(-1, 1), y)
# area = area.reshape(-1,1)
# price = price.reshape(-1,1)
# reg.fit(df[['area']],df.price)
reg.fit(area,price)
```

```python
# reg.predict(3300)
reg.predict(np.array([3300]).reshape(1, 1))
```

```python
reg.predict(np.array([4000]).reshape(1, 1))
```

```python
%matplotlib inline
plt.title("Price Prediction by area",fontsize=20)
plt.xlabel("Area",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(area,reg.predict(area),color='blue')
plt.show()
```
