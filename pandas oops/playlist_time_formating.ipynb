```python
from datetime import datetime

date_str = '09-01-2023 09:30:00'
date_obj = datetime.strptime(date_str, '%m-%d-%Y %H:%M:%S')
start_time = list()
end_time = list()
start_time.append(date_obj)
for d in df['Duration']:
    time_spt = d.split(':')
    time_spt1=time_spt[1].split('.')
    for_end_time = start_time[-1]
    date_object = datetime.strptime(str(for_end_time), '%Y-%m-%d %H:%M:%S')
    result_end_time = date_object + timedelta(minutes=int(time_spt[0]),seconds=int(time_spt1[0]))
    end_time.append(str(result_end_time))
    for_start_time = end_time[-1]
    start_time.append(for_start_time)
print(start_time)
```
