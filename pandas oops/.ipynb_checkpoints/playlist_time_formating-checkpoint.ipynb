```python
from datetime import datetime, timedelta

now = datetime.now()
print(now)  # 👉️ 2023-02-19 05:33:06.313209

result = now + timedelta(minutes=30)
print(result)  # 👉️ 2023-02-19 05:58:06.313209

# ✅ only get updated time
print(result.time())
```

```python
from datetime import datetime

date_str = '09-01-2023 09:30:00'
date_object = datetime.strptime(date_str, '%m-%d-%Y %H:%M:%S')
start_time = list()
end_time = list()
stat_time.append(date_object)
for d in df['Duration']:
    time_spt = d.split(':')
    for_end_time = start_time[-1]
    date_object = datetime.strptime(for_end_time, '%m-%d-%Y %H:%M:%S')
    result_end_time = date_object + timedelta(minutes=int(time_spt[0]),seconds=int(time_spt[1]))
    end_time.append(result_end_time)
    for_start_time = end_time[-1]
    stat_time.append(for_start_time)
print(start_time,'---',end_time)
```
